```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,4,6,8,10,12,14,16,18]

plt.plot(x,y,marker="*", color="green")
plt.title("Example")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x=["red","green","blue","orange"]
y=[5,19,2,4]
plt.bar(x,y, color="red")
plt.xlabel("color")
plt.ylabel("count")
plt.show()
```

```python
x=["red","green","blue","orange"]
y=[5,19,2,4]

plt.pie(y,labels=x, autopct="%1.1f%%",colors=["red","green","blue","orange"])
plt.title("Test")
plt.show()
```
