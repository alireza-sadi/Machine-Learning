```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4])
y = np.array([1,3,5,7])
```

```python
#reshape X
x= x.reshape(-1 , 1)
x
```

```python
# plot

plt.plot(x,y , "o")
plt.show()
```

```python
#model

#import model
from sklearn.linear_model import LinearRegression

#create model
model = LinearRegression()
model.fit(x,y)
```

```python
# y = m*x + c 

m = model.coef_[0]
c = model.intercept_

print (m,c)
```

```python
#plot predict
plt.plot(x,y , "o")
plt.plot(x, model.predict(x))
plt.show()
```

```python
# predict x=5
result = model.predict([[5]])
result
```
